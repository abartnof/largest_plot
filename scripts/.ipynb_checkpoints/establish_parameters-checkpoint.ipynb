## Easternly movement

```python
# Haversine part 1
# Holding y axis static (very North), move east. What is the relationship between decimal degrees and meters?

longitude_input_array = np.linspace(0.000028, 0.000029, num=1000000)
longitude_dict_north = {}

for d in longitude_input_array:
    start = (gdf.bounds['maxy'][0], gdf.bounds['maxx'][0]) # (lat, lon)
    end = (gdf.bounds['maxy'][0], gdf.bounds['maxx'][0] + d)
    m = haversine(start, end, Unit.METERS)
    longitude_dict_north[d] = m
    if m > 1:
        break

print('1 meter longitude is: ', d)
print('Real distance in meters is: ', m)
```

```python
# Haversine part 2
# Holding y axis static (very South), move east. What is the relationship between decimal degrees and meters?

longitude_input_array = np.linspace(0.00000001, 0.000029, num=100000)
longitude_dict_south = {}

for d in longitude_input_array:
    start = (gdf.bounds['miny'][0], gdf.bounds['maxx'][0]) # (lat, lon)
    end = (gdf.bounds['miny'][0], gdf.bounds['maxx'][0] + d)
    m = haversine(start, end, Unit.METERS)
    longitude_dict_south[d] = m
    if m > 1:
        break
        
print('1 meter longitude is: ', d)
print('Real distance in meters is: ', m)
```

### Counterfactuals
What if we impose the wrong decimal degree factors?

```python
# Counterfactual 1: impose the very south factor on the very north location
# Our resolution outperforms 1 meter

d = 9.291531215312152e-06
start = (gdf.bounds['maxy'][0], gdf.bounds['maxx'][0]) # (lat, lon)
end = (gdf.bounds['maxy'][0], gdf.bounds['maxx'][0] + d)
m = haversine(start, end, Unit.METERS)
print(m)
```

```python
# Counterfactual 2: impose the very north factor on the very south location
# Our resolution underperforms 1 meter

d = 2.8126270126270125e-05
start = (gdf.bounds['miny'][0], gdf.bounds['maxx'][0]) # (lat, lon)
end = (gdf.bounds['miny'][0], gdf.bounds['maxx'][0] + d)
m = haversine(start, end, Unit.METERS)
print(m)
```

conclusion: we will use 9.291531215312152e-06 as our change in longitude, irregardless of location. Worst case scenario, we're being too cautious

## Northerly movement

```python
# Haversine part 3
# Holding X axis static (very East), move north. What is the relationship between decimal degrees and meters?

latitude_input_array = np.linspace(0.00000001, 0.1, num=100000000)
latitude_dict_south = {}

for d in latitude_input_array:
    start = (gdf.bounds['miny'][0], gdf.bounds['maxx'][0]) # (lat, lon)
    end = (gdf.bounds['miny'][0] + d, gdf.bounds['maxx'][0])
    m = haversine(start, end, Unit.METERS)
    longitude_dict_south[d] = m
    if m > 1:
        break
        
print('1 meter latitude is: ', d)
print('Real distance in meters is: ', m)
```

```python
# Haversine part 4: difference in latitude (Min longitude)
# Holding X axis static (very West), move north. What is the relationship between decimal degrees and meters?

latitude_input_array = np.linspace(0.00000001, 0.1, num=100000000)
latitude_dict_south = {}

for d in latitude_input_array:
    start = (gdf.bounds['miny'][0], gdf.bounds['minx'][0]) # (lat, lon)
    end = (gdf.bounds['miny'][0] + d, gdf.bounds['minx'][0])
    m = haversine(start, end, Unit.METERS)
    longitude_dict_south[d] = m
    if m > 1:
        break
        
print('1 meter latitude is: ', d)
print('Real distance in meters is: ', m)
```

# Establish dummy files
## Chunk 1: x coordinates

We'll want 10 output files total. So: 
- Divide len(xx) by 10. This is the size of each output file.
- Assign the first output file an integer, 0. Repeat this integer until we have enough x values, and then increase the integer by 1, and do the same for the next batch of x values, until all x values have been assigned one of ten integers, flagging the file they'll output to.

```python
len_xx = len(np.arange(start=-125, stop=-66, step=9.291531215312152e-06))
num_records_per_chunk_x = round(len_xx / 10) 
num_records_per_chunk_x / 1e3
```

```python
chunk_array_x = np.zeros(len_xx)
chunk_num = 0
counter = 0
for i in np.arange(len_xx):
    chunk_array_x[i] = int(chunk_num)
    counter = int(counter + 1)
    is_auspicious = counter % int(num_records_per_chunk_x) == 0
    if is_auspicious:
        chunk_num = int(chunk_num + 1)
chunk_array_x
```

```python
# QC- size of each chunk
np.unique(chunk_array_x, return_counts=True)
```

## Chunk 2: y coordinates

```python
len_yy = len(np.arange(start=24, stop=50, step=8.993999191439993e-06))
num_records_per_chunk_y = round(len_yy / 10)  # num_records_per_chunk_x
num_records_per_chunk_y / 1e3
```

```python
chunk_array_y = np.zeros(len_yy)
chunk_num = 0
counter = 0
for i in np.arange(len_yy):
    chunk_array_y[i] = int(chunk_num)
    counter = int(counter + 1)
    is_auspicious = counter % int(num_records_per_chunk_y) == 0
    if is_auspicious:
        chunk_num = int(chunk_num + 1)
print(chunk_array_y)
```

```python
# QC- size of each chunk
np.unique(chunk_array_y, return_counts=True)
```

## Visualize grid breaks

```python
xx = np.arange(start=-125, stop=-66, step=9.291531215312152e-06)
yy = np.arange(start=24, stop=50, step=8.993999191439993e-06)
```

```python
plot_x_values = pd.DataFrame({'x':xx, 'chunk':chunk_array_x}).groupby('chunk').first()['x'].values
```

```python
plot_y_values = pd.DataFrame({'y':yy, 'chunk':chunk_array_y}).groupby('chunk').first()['y'].values
```

```python
fig, ax = plt.subplots(figsize=(10,8))
xlim = ([-125, -66])  #Longitude
ylim = ([24, 50])  #Latitude

ax.set_xlim(xlim)
ax.set_ylim(ylim)

gdf.plot(ax=ax)
plt.vlines(x=plot_x_values, ymin=min(ylim), ymax=max(ylim), colors='black', linestyles='dashed')
plt.hlines(y=plot_y_values, xmin=min(xlim), xmax=max(xlim), colors='black', linestyles='dashed')
plt.xticks(ticks=plot_x_values, labels=np.arange(10))
plt.yticks(ticks=plot_y_values, labels=np.arange(10))
plt.show()
```
