Overlay linear water features

NB- use a buffer width of 50m for water features

```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
# from shapely.geometry import Point, box, LineString
from shapely import intersection
from tqdm.notebook import tqdm
import glob

target_epsg = 6350  # https://epsg.org/crs_6350/NAD83-2011-Conus-Albers.html?sessionkey=nb47agmo4r
```

```python
fn_aoi = '/Volumes/Extreme SSD/largest_plots/clean_data/aoi.parquet'
fn_water_linear_overlay = '/Volumes/Extreme SSD/largest_plots/clean_data/water_linear_overlay.parquet'
dir_water_linear = '/Volumes/Extreme SSD/largest_plots/raw_data/linear_water/*.zip'
fn_list_water_linear = glob.glob(dir_water_linear)
```

```python
Aoi = gpd.read_parquet(fn_aoi).set_geometry('aoi').set_crs(epsg=target_epsg)
```

Iterate through each AOI x each water linear file.
- Keep any rows that overlap at all with the AOI
- Each AOI's results will be stored as a list of geopandas tables.

```python
intersection_list = []
for fn_water_linear in tqdm(fn_list_water_linear):
    WaterObservation = gpd.read_file(fn_water_linear)
    WaterObservation = WaterObservation[['geometry']]
    WaterObservation = WaterObservation.rename(columns={'geometry':'water_linear_geometry'})
    WaterObservation = WaterObservation.set_geometry('water_linear_geometry').to_crs(epsg=target_epsg)
    WaterObservation['water_linear_geometry'] = WaterObservation['water_linear_geometry'].buffer(distance=25, cap_style='round')
    Intersection = gpd.sjoin(left_df=WaterObservation, right_df=Aoi, how='inner', predicate='intersects', rsuffix='aoi')
    
    if len(Intersection) > 0:
        intersection_list.append(Intersection)
    else:
        pass
```

```python
Intersection = pd.concat(intersection_list).reset_index(drop=True)
```

```python
Aoi.loc[[i]].bounds
```

```python
overlay_dict = {}
for i in tqdm(Aoi.index):
    aoi_mask = Aoi.loc[[i]].bounds.values.tolist()[0]
    SubsetOfIntersection = Intersection.loc[Intersection.index_aoi == i]
    local_overlay = gpd.clip(gdf=SubsetOfIntersection, mask=aoi_mask, keep_geom_type=True).dissolve()
    overlay_dict[i] = local_overlay
```

```python
WaterLinearOverlay = pd.concat(overlay_dict.values())
WaterLinearOverlay = WaterLinearOverlay.set_index('index_aoi')
WaterLinearOverlay.index.names = ['index']
WaterLinearOverlay.head()
```

```python
# QC
i = 100
fig, ax = plt.subplots(figsize=(5, 5))
Aoi.loc[[i]].plot(ax=ax, color='black')
overlay_dict[i].plot(ax=ax)
plt.show()
```

```python
set(WaterLinearOverlay.index.values) == set(Aoi.index.values)
```

# Save to disk

```python
WaterLinearOverlay.to_parquet(fn_water_linear_overlay)
```
